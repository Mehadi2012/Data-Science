# Matplotlib Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine = pd.read_csv('/content/sample_data/whitewines.csv')
wine.head()
```

```python
figr,ax = plt.subplots()
ax.hist(wine['quality']) # Plot Histogram
ax.set_title('wine review scores') # Set Title
ax.set_xlabel('quality')  # Set X label. if we put another name here, it wont change any context of the function.
ax.set_ylabel('Frequency')   # Set Y label
```

```python
wine['quality'].plot.hist()   # another way of plotting histogram in pandas
```

# Density Plot in Pandas

```python
wine['quality'].plot(kind = 'density', subplots = True, layout = (1,2),  sharex = True, figsize = (15,5));
```

# Density plot in seaborn

```python
sns.distplot(wine['quality'], hist = False, rug = True);
```

```python
sns.kdeplot(wine['quality'])
plt.show()
```

# Box plot in seaborn

```python
sns.boxplot(x = 'quality', data = wine)  # Here x means we are displaying the plot in x axis
```

```python
sns.boxplot(y = 'quality', data = wine)  # Here y means we are displaying the plot in x axis
```

# Bar plot in seaborn

```python
sns.countplot(x = 'quality', data = wine)
```

# Scatter plot in seaborn: Numerical vs Numerical

```python
sns.scatterplot(x ='alcohol', y = 'pH', data = wine)
```

## Categorical vs Numerical

```python
sns.scatterplot(x ='quality', y = 'pH', data = wine)
```
